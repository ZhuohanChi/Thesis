```python
import pandas as pd

# Load NICU_30 data
nicu_30 = pd.read_csv(r"../data/final/nicu_30.csv")

def calculate_missing_rate(data):
    """
    Calculate and display the missing rate for each column in the DataFrame.

    Args:
        data (pd.DataFrame): Input DataFrame.

    Returns:
        pd.DataFrame: DataFrame showing column names and their missing rates.
    """
    missing_rate = data.isnull().mean() * 100
    missing_df = pd.DataFrame({
        "Column": missing_rate.index,
        "Missing Rate (%)": missing_rate.values
    }).sort_values(by="Column", ascending=True).reset_index(drop=True)
    pd.set_option('display.max_rows', None)
    print(missing_df)
    return missing_df

# Run the missing rate calculation on NICU_30 dataset
calculate_missing_rate(nicu_30)
```
