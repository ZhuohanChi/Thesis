```python
import pandas as pd
```

```python
nicu_120 = pd.read_csv(r"../data/final/nicu_120.csv")
```

```python
mean_features = [col for col in nicu_120.columns if "mean" in col]


demographics_table = pd.DataFrame()

demographics_table['total_mean'] = nicu_120[mean_features].mean()
demographics_table['total_std'] = nicu_120[mean_features].std()
demographics_table['is_infected_mean'] = nicu_120[nicu_120['is_infected'] == 1][mean_features].mean()
demographics_table['is_infected_std'] = nicu_120[nicu_120['is_infected'] == 1][mean_features].std()
demographics_table['not_infected_mean'] = nicu_120[nicu_120['is_infected'] == 0][mean_features].mean()
demographics_table['not_infected_std'] = nicu_120[nicu_120['is_infected'] == 0][mean_features].std()


demographics_table.to_csv(r"../data/final/demographics_table.csv", index=True)


print(demographics_table.head())
```
