```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.metrics import (
    roc_curve, auc,
    accuracy_score, precision_score, recall_score, f1_score, make_scorer
)

# ===== å¸¸ç”¨åˆ†ç±»å™¨ =====
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.ensemble import (
    GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier,
    ExtraTreesClassifier, BaggingClassifier, HistGradientBoostingClassifier
)

# å¦‚æžœæ²¡å®‰è£…ï¼Œè¯·å…ˆ pip install xgboost lightgbm catboost
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier


def run_model_selection(data: pd.DataFrame, target: str, id_column: str,
                        n_splits: int = 5, random_state: int = 42):
    """
    ä¸Žä¹‹å‰ç›¸åŒï¼šäº¤å‰éªŒè¯è¯„ä¼°å¹¶è¿”å›ž F1-score æœ€ä½³æ¨¡åž‹ã€‚
    """
    row_missing_ratio = data.isnull().sum(axis=1) / data.shape[1]
    data = data.loc[row_missing_ratio <= 0.75].copy()

    X = data.drop(columns=[target, id_column])
    y = data[target]

    imputer = IterativeImputer(max_iter=12, random_state=random_state)
    X_imputed = imputer.fit_transform(X)

    models = {
        "Decision Tree": DecisionTreeClassifier(),
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Naive Bayes": GaussianNB(),
        "KNN": KNeighborsClassifier(),
        "Multilayer Perceptron": MLPClassifier(max_iter=1000),
        "SVM": SVC(probability=True),
        "Gradient Boosting": GradientBoostingClassifier(),
        "AdaBoost": AdaBoostClassifier(),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss"),
        "Random Forest": RandomForestClassifier(),
        "LightGBM": LGBMClassifier(),
        "Extra Trees": ExtraTreesClassifier(),
        "CatBoost": CatBoostClassifier(verbose=0),
        "Bagging Classifier": BaggingClassifier(),
        "HistGradientBoosting": HistGradientBoostingClassifier(),
    }

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    results, best_model_name, best_f1 = [], None, 0.0
    scoring = {
        "accuracy": make_scorer(accuracy_score),
        "precision": make_scorer(precision_score, average="weighted"),
        "recall": make_scorer(recall_score, average="weighted"),
        "f1": make_scorer(f1_score, average="weighted"),
    }

    for name, model in models.items():
        acc = cross_val_score(model, X_imputed, y, cv=skf, scoring=scoring["accuracy"]).mean()
        pre = cross_val_score(model, X_imputed, y, cv=skf, scoring=scoring["precision"]).mean()
        rec = cross_val_score(model, X_imputed, y, cv=skf, scoring=scoring["recall"]).mean()
        f1  = cross_val_score(model, X_imputed, y, cv=skf, scoring=scoring["f1"]).mean()

        results.append([name, acc, pre, rec, f1])
        if f1 > best_f1:
            best_f1, best_model_name = f1, name

    print("\n===== Model Evaluation Results =====")
    print(pd.DataFrame(results, columns=["Model","Accuracy","Precision","Recall","F1"]).to_string(index=False))
    print(f"\n===== Best Model (F1) =====\n{best_model_name}  |  F1 = {best_f1:.4f}")

    best_model = models[best_model_name].fit(X_imputed, y)
    return best_model


def plot_all_models_roc(data: pd.DataFrame, target: str, id_column: str,
                        n_splits: int = 5, random_state: int = 42, top_k: int = 5):
    """
    ä»…ç»˜åˆ¶ AUC æœ€é«˜çš„å‰ top_k ä¸ªæ¨¡åž‹çš„ ROC æ›²çº¿ã€‚
    """
    row_missing_ratio = data.isnull().sum(axis=1) / data.shape[1]
    data = data.loc[row_missing_ratio <= 0.75].copy()

    X = data.drop(columns=[target, id_column])
    y = data[target]

    imputer = IterativeImputer(max_iter=12, random_state=random_state)
    X_imputed = imputer.fit_transform(X)

    models = {
        "Decision Tree": DecisionTreeClassifier(),
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Naive Bayes": GaussianNB(),
        "KNN": KNeighborsClassifier(),
        "Multilayer Perceptron": MLPClassifier(max_iter=1000),
        "SVM": SVC(probability=True),
        "Gradient Boosting": GradientBoostingClassifier(),
        "AdaBoost": AdaBoostClassifier(),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss"),
        "Random Forest": RandomForestClassifier(),
        "LightGBM": LGBMClassifier(),
        "Extra Trees": ExtraTreesClassifier(),
        "CatBoost": CatBoostClassifier(verbose=0),
        "Bagging Classifier": BaggingClassifier(),
        "HistGradientBoosting": HistGradientBoostingClassifier(),
    }

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    # å…ˆè®¡ç®—æ‰€æœ‰æ¨¡åž‹çš„ AUC
    auc_dict = {}
    roc_dict = {}
    for name, model in models.items():
        y_proba = cross_val_predict(model, X_imputed, y, cv=skf, method="predict_proba")
        fpr, tpr, _ = roc_curve(y, y_proba[:, 1])
        roc_auc = auc(fpr, tpr)
        auc_dict[name] = roc_auc
        roc_dict[name] = (fpr, tpr)

    # é€‰å‡º AUC æœ€é«˜çš„ top_k ä¸ªæ¨¡åž‹
    top_models = sorted(auc_dict.items(), key=lambda x: x[1], reverse=True)[:top_k]

    # æ‰“å° Topâ€‘k AUC
    print(f"\n===== Top {top_k} Models by AUC =====")
    for rank, (name, auc_val) in enumerate(top_models, 1):
        print(f"{rank}. {name:25s}  AUC = {auc_val:.4f}")

    # ç»˜å›¾
    plt.figure(figsize=(10, 8))
    for name, auc_val in top_models:
        fpr, tpr = roc_dict[name]
        plt.plot(fpr, tpr, label=f"{name} (AUC = {auc_val:.3f})")

    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves (Top 5 Models)")
    plt.legend(loc="lower right")
    plt.show()


def main(data: pd.DataFrame, target: str, id_column: str,
         n_splits: int = 5, random_state: int = 42):
    print(">>> Running Model Selection ...")
    best_model = run_model_selection(data, target, id_column, n_splits, random_state)

    print("\n>>> Plotting ROC Curves for Topâ€‘5 Models ...")
    plot_all_models_roc(data, target, id_column, n_splits, random_state, top_k=5)

    print("\n>>> Done.")
    return best_model


if __name__ == "__main__":
    nicu_30 = pd.read_csv(r"../data/final/nicu_30.csv")
    final_model = main(
        data=nicu_30,
        target="is_infected",
        id_column="SUBJECT_ID",
        n_splits=5,
        random_state=42
    )
```
