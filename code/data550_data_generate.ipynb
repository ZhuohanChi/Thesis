```python
import pandas as pd
```

```python
nicu_30 = pd.read_csv(r"../data/final/nicu_30.csv")
nicu_120 = pd.read_csv(r"../data/final/nicu_120.csv")
```

```python
nicu_30.info()
```

```python

df_filtered = nicu_30.loc[:, ~nicu_30.columns.str.contains("min|max", case=False)]
df_filtered = df_filtered.rename(columns=lambda x: x.replace("_mean_30", ""))

df_filtered.to_csv("../data/final/nicu_30_mean.csv", index=False)
```
